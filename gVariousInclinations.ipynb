```python
mean=np.mean(df["Absolute acceleration (m/s^2)"])
std=np.std(df["Absolute acceleration (m/s^2)"],ddof=1)
err_mean=std/np.sqrt(len(df["Absolute acceleration (m/s^2)"]))
print("acceleration=%3.3f ¬± %3.3f m/s^2"%(mean,err_mean))
print("acceleration=%3.2f ¬± %3.2f m/s^2"%(mean,std))
```

```python
mean_10=np.zeros(1000)
for i in range(1000):
    mean_10[i]=df["Absolute acceleration (m/s^2)"].sample(n=10).mean()
```

```python
import matplotlib.pyplot as pl
pl.hist(mean_10,bins=75,range=(9.75,9.9));
pl.grid();
pl.xlabel('Acceleration mean of 10 samples (m/s^2)');
```
