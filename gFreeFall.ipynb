```python
import pandas as pd
import matplotlib as mp
import numpy as np
```

```python
df=pd.read_csv("FreeFall_59cm/Raw Data.csv")
```

```python
df.describe()
```

```python
plt=df.plot(x="Time (s)",title="Acceleration vs time",style='o-',ms=4,mfc='none',grid=True);
plt.set_ylabel("Absolute acceleration (m/s^2)");
```

```python
df_fall=df.loc[(df['Absolute acceleration (m/s^2)']<10) & (df['Time (s)']<0.93)]
```

```python
df_fall.describe()
```

```python
plt=df_fall.plot(x="Time (s)",title="Acceleration vs time",style='o-',ms=4,mfc='none',grid=True);
plt.set_ylabel("Absolute acceleration (m/s^2)");
```

```python
df_start_fall=df_fall.loc[(df['Time (s)']<0.6)]
```

```python
plt=df_start_fall.plot(x="Time (s)",title="Acceleration vs time",style='o-',ms=4,mfc='none',grid=True)
plt.set_xlim(0.52,0.58)
plt.set_ylabel("Absolute acceleration (m/s^2)")
plt.set_xticks(np.arange(0.52, 0.58, 0.005));
```

```python
df_end_fall=df_fall.loc[(df['Time (s)']>0.8) & (df['Time (s)']<0.9)]
```

```python
plt=df_end_fall.plot(x="Time (s)",title="Acceleration vs time",style='o-',ms=4,mfc='none',grid=True)
plt.set_xlim(0.85,0.91)
plt.set_ylabel("Absolute acceleration (m/s^2)")
plt.set_xticks(np.arange(0.85, 0.91, 0.005));
```
